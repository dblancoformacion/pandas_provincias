# Titanic

```python
import pymysql
con = pymysql.connect('localhost','root','','provincias')
conn = con.cursor()
```

```python
conn.execute("""
    SELECT * FROM titanic;
""")
rs = conn.fetchall()
```

```python
import pandas as pd
df = pd.DataFrame(rs)
df.columns=[
    'PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'
]
df
```

### Nombre de los 5 primeros pasajeros

### Listado de los pasajeros de primera clase

```python
conn.execute("""
    SELECT * FROM titanic
      WHERE PClass=1;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df[ df.Pclass==1 ]
```

### Pasajeros de 1Âª y 2Âª clase

```python
conn.execute("""
    SELECT * FROM titanic
      WHERE Pclass=1 OR Pclass=2;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df[ df.Pclass.isin([1,2]) ]
```

```python
conn.execute("""
    SELECT * FROM titanic
      WHERE Pclass IN (1,2);
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df[ (df.Pclass==1) | (df.Pclass==2) ]
```

### Nombres de los varones de 1Âª clase

```python
conn.execute("""
    SELECT * FROM titanic
      WHERE Pclass=1 AND Sex='male';
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df[ (df.Pclass==1) & (df.Sex=='male') ]
```

### NÂº de pasajeros por clase y gÃ©nero

```python
conn.execute("""
    SELECT Pclass,Sex,COUNT(*) n 
      FROM titanic
      GROUP BY 1,2;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df.groupby(['Pclass','Sex']).count().PassengerId
```

```python
df.groupby(['Pclass','Sex']).size()
```

### NÂº de pasajeros por clase y gÃ©nero, ordenados por nÃºmero de pasajeros descendentemente

```python
conn.execute("""
    SELECT Pclass,Sex,COUNT(*) n 
      FROM titanic
      GROUP BY 1,2
      ORDER BY 3 DESC;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df.groupby(['Pclass','Sex']).size().sort_values(ascending=False)
```

### NÂº de pasajeros por clase y gÃ©nero, ordenados por clase y gÃ©nero

```python
conn.execute("""
    SELECT Pclass,Sex,COUNT(*) n 
      FROM titanic
      GROUP BY 1,2
      ORDER BY 1,2;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df.groupby(['Pclass','Sex']).size().sort_index()
```

### Listado de pasajeros, clase y gÃ©nero ordenados por clase descendente y nombre

```python
conn.execute("""
    SELECT Name,Pclass,Sex
      FROM titanic
      ORDER BY 2 DESC, 1;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df.sort_values(['Pclass','Name'],ascending=[False,True])[['Name','Pclass','Sex']]
```

### RecaudaciÃ³n por clase y gÃ©nero

```python
conn.execute("""
    SELECT Pclass,Sex,SUM(fare) usd
      FROM titanic
      GROUP BY 1,2;
""")
rs = conn.fetchall()
for i in rs: print(i)
```

```python
df.groupby(['Pclass','Sex']).sum()['Fare']
```
